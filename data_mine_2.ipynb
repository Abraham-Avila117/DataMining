```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# use "plt.close('all')" if you have plots taking up storage

# Read csv file 
data_path = './CSV/adults.csv'
adult_data = pd.read_csv(data_path)
```

```python
# display 10 records
print(adult_data.columns)
first_10_records = adult_data.head(10)
print(first_10_records)
```

```python
# calculate correlation matrix
correlation_matrix = adult_data.corr()
```

```python
# heat map
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmep")
plt.show()
```

```python
# identify target 
target_class_column = 'class'
categorical_columns = adult_data.select_dtypes(include=['object']).columns
```

```python
# create bar charts for the identified target
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=adult_data, x=column, hue=target_class_column, palette='Set2')
    plt.title(f'Bar Chart of {column} vs. {target_class_column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=60)
    plt.legend(title=target_class_column, loc='upper right')
    plt.show()
```
